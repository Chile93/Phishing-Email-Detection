## Exploratory Data Analysis (EDA)

```python
# Creating a bar chart
# Count the occurrences of each E-mail type. 
email_type_counts = phishing_df['Email Type'].value_counts()

# Create a list of unique email types
unique_email_types = email_type_counts.index.tolist()

# Define a custom color map 
color_map = {
    'Phishing Email': 'red',
    'Safe Email': 'green',}

# Map the colors to each email type
colors = [color_map.get(email_type, 'gray') for email_type in unique_email_types]

# Create the bar chart with custom colors
plt.figure(figsize=(7, 5))
plt.bar(unique_email_types, email_type_counts, color=colors)
plt.ylabel('Count')
plt.title('Count of Email Types')
plt.xticks(rotation=45)

# Show the chart
plt.tight_layout()
plt.show()
```

From the bar chart above we have imbalance data between the Safe Email and the Phishing Email. 
To handle this, we implement undersampling technique.
Undersampling: We reduce the number of instances in the majority class to match the minority class, effectively balancing the dataset.

### Undersampling the Majority Dataset

```python
# implementing the undersampling technique 
Safe_Email = phishing_df[phishing_df["Email Type"]== "Safe Email"]
Phishing_Email = phishing_df[phishing_df["Email Type"]== "Phishing Email"]
Safe_Email = Safe_Email.sample(Phishing_Email.shape[0])
```

```python
# check the shape again 
Safe_Email.shape,Phishing_Email.shape
```

```python
# concatenate the balanced dataset
phishing_bd = pd.concat([Safe_Email, Phishing_Email], ignore_index = True)
phishing_bd = phishing_bd.sample(frac=1, random_state=42).reset_index(drop=True)
phishing_bd.head()
```

Some rows in the *Email Text* have the word *empty*, so I will remove those columns

```python
# Remove rows with 'empty' in the 'Email Text' column
phishing_bd = phishing_bd[~phishing_bd['Email Text'].str.contains('empty')]
```

## Data Splitting and Vectorization

```python
# view the tokenized text
phishing_bd.head()
```

```python
# Drop the 'Email Text' and 'ID' columns
phishing_bd.drop(columns=['ID'], inplace=True)
```

Categorising 'Safe Email' as 0, and  'Phishing Email' as 1

```python
# Mapping the categorical values to numeric factors
phishing_bd['Email Type'] = phishing_bd['Email Type'].replace({'Safe Email': 0, 'Phishing Email': 1})
```

```python
# view the head
phishing_bd.head()
```

```python
# Defining our features and target as X and y respectively
X = phishing_bd["Email Text"]
y = phishing_bd["Email Type"]
```

```python
# splitting the data with train_test_split in the ratio of 7:3 and setting the random state to 42
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42

)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Initialize TfidfVectorizer
vectorizer = TfidfVectorizer(sublinear_tf=True, ngram_range=(1, 3), max_features=10000)

# Fit and transform the 'X_train' and 'X_test' separately
X_train_vd = vectorizer.fit_transform(X_train)
X_test_vd = vectorizer.transform(X_test)
```

## Machine Learning Models

### Train Logistic Regression 

```python
# Cross validate model with Kfold stratified cross val
kfold = StratifiedKFold(n_splits=10)
kfold
```

```python
#creating metric function 
def metrics_score(actual, predicted):
    print(classification_report(actual, predicted))
    cm = confusion_matrix(actual, predicted)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True,  
                fmt='.2f', xticklabels=['Safe Mail', 'Phishing Mail'], 
                yticklabels=['Safe Mail', 'Phishing Mail'], 
                cmap="crest")
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show();
```

```python
# Define parameter grids for each model
C = np.logspace(-2, 1, 20)
param_grid_lr = {'penalty': ['l1', 'l2'], 'C': C}

# Instantiate models
lr = LogisticRegression(max_iter=1000)  # high max_iter for convergence

# Use GridSearchCV with a larger k for k-fold cross-validation
grid_lr = GridSearchCV(lr, param_grid_lr, cv=kfold, scoring='accuracy', n_jobs=4, verbose=1)

# Fit models
model_lr = grid_lr.fit(X_train_vd, y_train)
lr_best = model_lr.best_estimator_

# Get best parameters and best scores
print("Logistic Regression:", model_lr.best_params_, model_lr.best_score_)

# View best hyperparameters
print('Best Penalty:', lr_best.get_params()['penalty'])
print('Best C:', lr_best.get_params()['C'])

# Best score
model_lr.best_score_
```

```python
#checking the performance on the test dataset
y_pred_test_lr = model_lr.predict(X_test_vd)
metrics_score(y_test, y_pred_test_lr)
```

```python
# ROC Curve
y_scores_lr = model_lr.predict_proba(X_test_vd)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_scores_lr)
auc = roc_auc_score(y_test, y_scores_lr)

plt.figure()
plt.plot(fpr, tpr, label=f'{"Logistic Regression"} (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title(f'ROC Curve - {"Logistic Regression"}')
plt.legend(loc='lower right')
plt.show()
```

```python
# Test the model with some custom email messages

message = ["Congratulations on your recent achievement! Well done."]
input_data_features = vectorizer.transform(message)
prediction = model_lr.predict(input_data_features)

if (prediction)[0] == 0:
    print("Safe Mail")
else:
    print("Phishing Mail")
```

## Deep Learning Models

### LSTM Model

```python
# Tokenize text
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train) 
X_train_seq = tokenizer.texts_to_sequences(X_train)
X_test_seq = tokenizer.texts_to_sequences(X_test)

# Pad sequences to a fixed length of 100
max_length = 100 
X_train_pad = pad_sequences(X_train_seq, maxlen=max_length)
X_test_pad = pad_sequences(X_test_seq, maxlen=max_length)

vocab_size = len(tokenizer.word_index) + 1  # Adding 1 for the reserved 0 index

# Define LSTM model
model = Sequential()
model.add(Embedding(input_dim=vocab_size, output_dim=100, input_length=max_length))
model.add(LSTM(units=128, return_sequences=True))
model.add(LSTM(units=64)) 
model.add(Dense(units=1, activation='sigmoid'))
optimizer = Adam(learning_rate=0.001) 

# Compile the model
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model_lstm = model.fit(X_train_pad, y_train, epochs=20, batch_size=32, validation_data=(X_test_pad, y_test))
```

```python
# Evaluate the model on test data
test_loss, test_accuracy = model_lstm.model.evaluate(X_test_pad, y_test)
print(f"Test Accuracy: {test_accuracy}")
```

```python
# Get the trained model from the variable 'model'
trained_model_lstm = model_lstm.model

# making predictions on the test dataset
y_pred_test_l = trained_model_lstm.predict(X_test_pad)
y_pred_test_l
```

```python
# Convert predictions to binary classes (0 or 1)
y_pred_classes_l = (y_pred_test_l > 0.5).astype(int) 
#checking the performance on the test dataset
metrics_score(y_test, y_pred_classes_l)
```

```python
# ROC Curve for LSTM
fpr, tpr, _ = roc_curve(y_test, y_pred_test_l)  # Use the scores directly for binary classification
auc = roc_auc_score(y_test, y_pred_test_l)

plt.figure()
plt.plot(fpr, tpr, label=f'{"LSTM"} (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - LSTM')
plt.legend(loc='lower right')
plt.show()
```

### Model Evaluation and Comparison

```python
# function to calculate metrics
def calculate_metrics(y_true, y_pred):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    auc = roc_auc_score(y_true, y_pred)
    return accuracy, precision, recall, f1, auc

# Calculate the metrics for each model
models = [
    ("Logistic Reg", y_pred_test_lr),
    ("Naive Bayes", y_pred_test_nvb),
    ("Random Ft", y_pred_test_rf),
    ("SVM", y_pred_test_svm),
    ("LSTM", y_pred_classes_l),
    ("GRU", y_pred_classes_g),
    ("CNN", y_pred_classes_c),
    ("DistilBert", y_pred_classes_db)
]

results = []
for model_name, y_pred in models:
    metrics = calculate_metrics(y_test, y_pred)
    results.append((model_name,) + metrics)

# put them in DataFrame
columns = ["Model", "Accuracy", "Precision", "Recall", "F1-Score", "AUC"]
df = pd.DataFrame(results, columns=columns)
df
```

```python
# Defining the models and metrics to plot from the dataframe (df)
models = df['Model']
metrics = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'AUC']

# Set the width of the bars
bar_width = 0.15

# Set the position of the bars on the x-axis
r = np.arange(len(models))

# Create the bar plot
plt.figure(figsize=(9, 6))
for i, metric in enumerate(metrics):
    plt.bar(r + (i * bar_width), df[metric], bar_width, label=metric)

# Add labels, title, and ticks
plt.xlabel('Models')
plt.ylabel('Metrics')
plt.title('Performance Metrics Comparison')
plt.xticks(r + bar_width * 2, models)
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
